```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, display

SAMPLE_RATE = 44100

audio_path = 'C:/Users/bryan/Documents/GitHub/Synth/exports/Acordeon2_predicted_[0.06535688].wav'
y, sr = librosa.load(audio_path, sr=SAMPLE_RATE)
y = np.array(y, dtype=np.float64)

S = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=256, n_mels=128)

S_dB = librosa.power_to_db(S, ref=np.max)

librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Espectrograma de Mel')
plt.tight_layout()
plt.show()

display(Audio(y, rate=SAMPLE_RATE))
```
